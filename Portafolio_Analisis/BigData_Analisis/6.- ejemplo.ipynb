```python
from pyspark import SparkContext
sc = SparkContext()
```

```python
rdd = sc.textFile("LaCelestina.txt")
```

```python
rdd.takeSample(False, 10)
```

```python
palabras = rdd.map(lambda l: l.split(" "))
```

```python
palabras.take(4)
```

```python
palabras = rdd.flatMap(lambda l: l.split(" "))
```

```python
palabras.take(4)
```

```python
palabras.takeSample(False,20)
```

```python
palabras = rdd.flatMap(lambda l: l.split(" ")).map(lambda s: s.strip(",.-_*ยก?ยก!\(\)\'\"/''")).countByValue()
```
